```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

# Q1 a: First get the stable exogeneous gradient

```python
C0=1
D=1 #diffusion coefficient: u_t=D*u_xx

L=10
h=1
T=20
tau=0.2*h**2 #tau<h**2!!!!!!! h=0.2, tau=0.005 

if T%tau==0:
    M=int(T//tau+1)
else:
    M=int(T//tau+2)
if L%h==0:
    N=int(L//h+1)
else:
    N=int(L//h+2)

u=np.zeros((M,N))
for i in range(M):
    u[i][0]=0.0
    u[i][-1]=C0
for j in range(1,N-1):
    u[0][j]=0.0

for i in range(1,M):
    for j in range(1,N-1):
        u[i][j]=u[i-1][j] +tau*D*(u[i-1][j-1]-2*u[i-1][j]+u[i-1][j+1])/h**2


        

for i in range(1,M):
    plt.plot([np.min([j*h,L]) for j in range(N)], u[i])
plt.xlabel('x, [0,L], 1D')
plt.ylabel('C(x)')
plt.title('Concentration of chemical after stabilization')
plt.show()
```

```python
class Consentration:
    def __init__(self,u,L,N,h,T,M,tau):
        mas_x=[np.min([j*h,L]) for j in range(N)]
        self.C=interpolate.interp1d(mas_x,u[-1])
    def getC(self,point):
        x,y=point[0],point[1]
        return self.C(x).flatten()[0] #C    x
    
MyConsentration=Consentration(u,L,N,h,T,M,tau)
MyConsentration.getC((6,1000)).flatten()[0]
```

# Q1b: Gillespie for imposed gradient

```python


class Graph1DImposedGradient:
    def __init__(self,GraphNodesLocations,AdjacencyMatrix,TumourSellsLocations):
        self.AmtGraphNodes=len(AdjacencyMatrix)
        self.GraphNodesLocations=GraphNodesLocations
        self.AdjacencyMatrix=AdjacencyMatrix
        self.AmtTumorSells=len(TumourSellsLocations)
        self.TumourSellsLocations=TumourSellsLocations
    def propensity_function(self,ci,cj):
        return np.exp(np.abs(ci-cj))
    def get_alphas(self,Cons):
        props=np.zeros_like(self.AdjacencyMatrix)
        for i in range(self.AmtGraphNodes):
            for j in range(self.AmtGraphNodes):
                ci=Cons.getC(self.GraphNodesLocations[i])
                cj=Cons.getC(self.GraphNodesLocations[j])
                props[i][j]=self.propensity_function(ci,cj)*self.AdjacencyMatrix[i][j]
        alphas=[]
        info_for_alphas=[]
        for tumor_sell_number,i in self.TumourSellsLocations.items():
            for j in range(self.AmtGraphNodes):
                if props[i][j]!=0:
                    alphas.append(props[i][j])
                    info_for_alphas.append((tumor_sell_number,i,j))
        return alphas,info_for_alphas
       
    def updateOneTumorSellLocation(self,tumor_sell_number,node_from,node_to):
        if self.TumourSellsLocations[tumor_sell_number]!=node_from:
            raise Exception('Node_from is corrupted!')
        else:
            self.TumourSellsLocations[tumor_sell_number]=node_to
    def getAmtGraphNodes(self):
        return self.AmtGraphNodes
    def getGraphNodesLocations(self):
        return self.GraphNodesLocations
    def getAdjacencyMatrix(self):
        return self.AdjacencyMatrix
    def getTumourSellsLocations(self):
        return self.TumourSellsLocations


K=100  #5 tumor sells      
print("N=",N, "K=",K)

MyAdjacencyMatrix=np.zeros((N,N)) #  1,  
for i in range(N):
    if (i+1<=N-1):
        MyAdjacencyMatrix[i][i+1]=1
    if (i-1>=0):
        MyAdjacencyMatrix[i][i-1]=1
print("MyAdjacencyMatrix=")
#print(MyAdjacencyMatrix)

MyGraph1DImposedGradient=Graph1DImposedGradient(GraphNodesLocations=[(g,0) for g in range(N)], 
                    AdjacencyMatrix=MyAdjacencyMatrix, TumourSellsLocations={g:0 for g in range(K)})

t=0
for s in range(2000): #7   
    alphas,info_for_alphas=MyGraph1DImposedGradient.get_alphas(MyConsentration)
    
    #find time of next reaction
    r1=np.random.random()
    tau=np.log(1/r1)/sum(alphas)
    t+=tau
    
    #find which reaction occurs
    ind=np.random.choice(a=[i for i in range(len(info_for_alphas))],p=np.array(alphas)/sum(alphas))
    tumor_sell_number,node_from,node_to=info_for_alphas[ind]
    #print("Step",s+1,"tumor_sell_number=",tumor_sell_number,"node_from=",node_from,"node_to=",node_to)
    MyGraph1DImposedGradient.updateOneTumorSellLocation(tumor_sell_number,node_from,node_to)
    
    positions,counts=np.unique(list(MyGraph1DImposedGradient.getTumourSellsLocations().values()), return_counts=True)
    positions=list(positions)
    counts=list(counts)
    for g in range(N):
        if g not in positions:
            positions.append(g)
            counts.append(0)
    #print(positions, counts)
    #print(sum(counts),tuple(np.array([1,0,0]) + (4+1)/K*(np.array((0,1,0)) - np.array([1,0,0]) )))
    #mycolors=[ tuple(np.array([1,0,0]) + (g+1)/K*np.array((0,1,0))) for g in counts]
    mysizes=[ 1 + (g)/(K-1)*500 for g in counts]
    #print("mycolors=",mycolors)
    plt.scatter([t for g in range(len(positions))], positions, s=mysizes,c='blue')
    print("After step",s+1,counts)
    
    

plt.plot()
```

# Q1c: Gillespie for self-imposed gradients

```python

class Graph1DSelfGeneratedGradient:
    def __init__(self,GraphNodesLocations,AdjacencyMatrix,TumourSellsLocations,Concentrations):
        self.AmtGraphNodes=len(AdjacencyMatrix)
        self.GraphNodesLocations=GraphNodesLocations
        self.AdjacencyMatrix=AdjacencyMatrix
        self.AmtTumorSells=len(TumourSellsLocations)
        self.TumourSellsLocations=TumourSellsLocations
        self.Concentrations=Concentrations
    def propensity_function(self,ci,cj):
        return np.exp(np.abs(ci-cj))
    def get_alphas(self):
        props=np.zeros_like(self.AdjacencyMatrix)
        for i in range(self.AmtGraphNodes):
            for j in range(self.AmtGraphNodes):
                ci=self.Concentrations[i]
                cj=self.Concentrations[j]
                props[i][j]=self.propensity_function(ci,cj)*self.AdjacencyMatrix[i][j]
        alphas=[]
        info_for_alphas=[]
        for tumor_sell_number,i in self.TumourSellsLocations.items():
            for j in range(self.AmtGraphNodes):
                if props[i][j]!=0:
                    alphas.append(props[i][j])
                    info_for_alphas.append((tumor_sell_number,i,j))
        return alphas,info_for_alphas
       
    def updateOneTumorSellLocation(self,tumor_sell_number,node_from,node_to):
        if self.TumourSellsLocations[tumor_sell_number]!=node_from:
            raise Exception('Node_from is corrupted!')
        else:
            self.TumourSellsLocations[tumor_sell_number]=node_to
    def updateConcentrations(self,new_concentrations):
        self.Concentrations=new_concentrations
    def getAmtGraphNodes(self):
        return self.AmtGraphNodes
    def getGraphNodesLocations(self):
        return self.GraphNodesLocations
    def getAdjacencyMatrix(self):
        return self.AdjacencyMatrix
    def getTumourSellsLocationsIdToCompartment(self):
        return self.TumourSellsLocations
    def getTumourSellsLocationsCompartmentToAmt(self):
        locations,counts=np.unique(list(self.getTumourSellsLocationsIdToCompartment().values()), return_counts=True)
        mydict={a:b for (a,b) in zip(locations, counts)}
        locs=[0 for g in range(self.AmtGraphNodes)]
        cnts=[0 for g in range(self.AmtGraphNodes)]
        for g in range(self.AmtGraphNodes):
            if g not in locations:
                cnts[g]=0
            else:
                cnts[g]=mydict[g]
        return cnts
    def getConcentrations(self):
        return self.Concentrations

N=11
K=5 #5 tumor sells      
print("N=",N, "K=",K)

MyAdjacencyMatrix=np.zeros((N,N)) #  1,  
for i in range(N):
    if (i+1<=N-1):
        MyAdjacencyMatrix[i][i+1]=1
    if (i-1>=0):
        MyAdjacencyMatrix[i][i-1]=1
print("MyAdjacencyMatrix=")
print(MyAdjacencyMatrix)

MyGraph1DSelfGeneratedGradient=Graph1DSelfGeneratedGradient(GraphNodesLocations=[(g,0) for g in range(N)], 
                    AdjacencyMatrix=MyAdjacencyMatrix, TumourSellsLocations={g:0 for g in range(K)},
                                                           Concentrations=[1.0 for g in range(N)])

#print(MyGraph1DSelfGeneratedGradient.getConcentrations())

vmax=1
kmax=10

t=0
for s in range(500): #7   
    
    alphas,info_for_alphas=MyGraph1DSelfGeneratedGradient.get_alphas()
    #print("alhpas=",alphas)
    
    #find time of next reaction
    r1=np.random.random()
    tau=np.log(1/r1)/sum(alphas)
    t+=tau
    
    
    #print("tau=",tau, "0.5*h**2=",0.5*h**2, "tau < 0.5*h**2=",tau < 0.5*h**2)
    mas_taus=[tau]
    if (tau >0.5*h**2):
        dtau=0.4*h*h
        mas_taus=[dtau for g in range(int(tau//dtau))]
        mas_taus.append(tau-sum(mas_taus))
        
    
    tumorSellCounts=MyGraph1DSelfGeneratedGradient.getTumourSellsLocationsCompartmentToAmt()
    
    for tau in mas_taus:
        cons=MyGraph1DSelfGeneratedGradient.getConcentrations()
        new_cons=[-1 for g in range(N)]
        for j in range(1,N-1):
            new_cons[j]=cons[j]+tau*(cons[j-1]-2*cons[j]+cons[j+1])/h**2 - tumorSellCounts[j]*vmax*cons[j]/(cons[j]+kmax)
        new_cons[0]=new_cons[1]
        new_cons[-1]=new_cons[-2]
        #print("new_cons=", new_cons)
        MyGraph1DSelfGeneratedGradient.updateConcentrations(new_cons)

    
    #find which reaction occurs
    alphas,info_for_alphas=MyGraph1DSelfGeneratedGradient.get_alphas()
    ind=np.random.choice(a=[i for i in range(len(info_for_alphas))],p=np.array(alphas)/sum(alphas))
    tumor_sell_number,node_from,node_to=info_for_alphas[ind]
    #print("Step",s+1, "ind=",ind,"tumor_sell_number=",tumor_sell_number,"node_from=",node_from,"node_to=",node_to)
    MyGraph1DSelfGeneratedGradient.updateOneTumorSellLocation(tumor_sell_number,node_from,node_to)
    
    positions=np.arange(N)
    counts=MyGraph1DSelfGeneratedGradient.getTumourSellsLocationsCompartmentToAmt()
    
    #print(positions, counts)
    #print(sum(counts),tuple(np.array([1,0,0]) + (4+1)/K*(np.array((0,1,0)) - np.array([1,0,0]) )))
    #mycolors=[ tuple(np.array([1,0,0]) + (g+1)/K*np.array((0,1,0))) for g in counts]
    mysizes=[ 1 + (g)/(K-1)*500 for g in counts]
    #print("mycolors=",mycolors)
    plt.scatter([t for g in range(len(positions))], positions, s=mysizes,c='blue')
    print("After step",s+1,counts)
    #print(MyGraph1DSelfGeneratedGradient.getTumourSellsLocationsIdToCompartment())
    #print('\n')
   
    
    

plt.plot()
```

# Q2a: both ends are dead ends

```python
K=10
N=15
MyAdjacencyMatrix=np.zeros((N,N)) #  1,  
for i in range(N):
    if (i==4):
        MyAdjacencyMatrix[4][5]=1
        MyAdjacencyMatrix[4][3]=1
        MyAdjacencyMatrix[4][10]=1
    elif (i==9):
        MyAdjacencyMatrix[9][8]=1
    elif (i==10):
        MyAdjacencyMatrix[10][4]=1
        MyAdjacencyMatrix[10][11]=1
    else:
        if (i+1<=N-1):
            MyAdjacencyMatrix[i][i+1]=1
        if (i-1>=0):
            MyAdjacencyMatrix[i][i-1]=1
print("MyAdjacencyMatrix=")
#print(MyAdjacencyMatrix)

MyGraph1DSelfGeneratedGradient=Graph1DSelfGeneratedGradient(GraphNodesLocations=[(g,0) for g in range(N)], 
                    AdjacencyMatrix=MyAdjacencyMatrix, TumourSellsLocations={g:0 for g in range(K)},
                                                           Concentrations=[1.0 for g in range(N)])

#print(MyGraph1DSelfGeneratedGradient.getConcentrations())

vmax=1
kmax=10

t=0
for s in range(200): #7   
    #print(MyGraph1DSelfGeneratedGradient.getTumourSellsLocationsIdToCompartment())
    alphas,info_for_alphas=MyGraph1DSelfGeneratedGradient.get_alphas()
    #print("alhpas=",alphas)
    #print("info_for_alphas=",info_for_alphas)
    
    #find time of next reaction
    r1=np.random.random()
    tau=np.log(1/r1)/sum(alphas)
    t+=tau
    
    #print("tau=",tau, "0.5*h**2=",0.5*h**2, "tau < 0.5*h**2=",tau < 0.5*h**2)
    mas_taus=[tau]
    if (tau >0.5*h**2):
        dtau=0.4*h*h
        mas_taus=[dtau for g in range(int(tau//dtau))]
        mas_taus.append(tau-sum(mas_taus))
    
    tumorSellCounts=MyGraph1DSelfGeneratedGradient.getTumourSellsLocationsCompartmentToAmt()
    
    for tau in mas_taus:
        cons=MyGraph1DSelfGeneratedGradient.getConcentrations()
        new_cons=[-1 for g in range(N)]
        for j in range(1,9):
            new_cons[j]=cons[j]+tau*(cons[j-1]-2*cons[j]+cons[j+1])/h**2 - tumorSellCounts[j]*vmax*cons[j]/(cons[j]+kmax)
        new_cons[9]=new_cons[10]
        new_cons[0]=new_cons[1]
        for j in range(11,14):
            new_cons[j]=cons[j]+tau*(cons[j-1]-2*cons[j]+cons[j+1])/h**2 - tumorSellCounts[j]*vmax*cons[j]/(cons[j]+kmax)
        new_cons[10]=cons[10]+tau*(cons[4]-2*cons[10]+cons[11])/h**2 - tumorSellCounts[10]*vmax*cons[10]/(cons[10]+kmax)
        new_cons[-1]=new_cons[-2]
        #print("new_cons=", new_cons)
        MyGraph1DSelfGeneratedGradient.updateConcentrations(new_cons)
    
    
    #find which reaction occurs
    alphas,info_for_alphas=MyGraph1DSelfGeneratedGradient.get_alphas()
    ind=np.random.choice(a=[i for i in range(len(info_for_alphas))],p=np.array(alphas)/sum(alphas))
    tumor_sell_number,node_from,node_to=info_for_alphas[ind]
    #print("Step",s+1, "ind=",ind,"tumor_sell_number=",tumor_sell_number,"node_from=",node_from,"node_to=",node_to)
    MyGraph1DSelfGeneratedGradient.updateOneTumorSellLocation(tumor_sell_number,node_from,node_to)
    
    positions=np.arange(N)
    counts=MyGraph1DSelfGeneratedGradient.getTumourSellsLocationsCompartmentToAmt()
    
    #print(positions, counts)
    #print(sum(counts),tuple(np.array([1,0,0]) + (4+1)/K*(np.array((0,1,0)) - np.array([1,0,0]) )))
    #mycolors=[ tuple(np.array([1,0,0]) + (g+1)/K*np.array((0,1,0))) for g in counts]
    mysizes=[ 1 + (g)/(K-1)*500 for g in counts]
    #print("mycolors=",mycolors)
    plt.scatter([t for g in range(len(positions))], positions, s=mysizes,c='blue')
    print("After step",s+1,counts)
    #print('\n')
   
    
    

plt.plot()
```
